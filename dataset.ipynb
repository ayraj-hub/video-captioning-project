```python
from datasets import load_dataset
import pandas as pd

# Load dataset
ds = load_dataset("friedrichor/MSR-VTT", "train_7k")

data = ds["train"]

expanded_rows = []

for item in data:
    video_id = item["video_id"]
    captions = item["caption"]   # 👈 NOTE: key is "caption", not "captions"
    
    for cap in captions:
        expanded_rows.append({
            "video_id": video_id,
            "caption": cap
        })

df = pd.DataFrame(expanded_rows)

print("Total rows:", len(df))
print("Unique videos:", df["video_id"].nunique())
print("Captions per video (example):")
print(df.groupby("video_id").size().head())
```

```python
import numpy as np

np.random.seed(42)

unique_videos = df["video_id"].unique()
selected_videos = np.random.choice(unique_videos, 2000, replace=False)

subset_df = df[df["video_id"].isin(selected_videos)]

print("Subset unique videos:", subset_df["video_id"].nunique())
print("Subset captions:", len(subset_df))
```

```python
video_folder = "/Users/ayraj/Desktop/project 2/TrainValVideo"

subset_df["video_path"] = subset_df["video_id"].apply(
    lambda x: f"{video_folder}/{x}.mp4"
)

subset_df.to_csv("msrvtt_train_2k_fullcaptions.csv", index=False)

print("2K dataset with full captions saved!")
```
